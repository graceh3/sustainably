### 0. Import all necessary libraries

```python
import numpy as np
import time
from bs4 import BeautifulSoup
import requests
import re
import pandas as pd
import urllib
import pandas as pd
import cv2
```

### 1. Get image link for all products

```python
urls = ["https://www.gunasthebrand.com/collections/shoulder-bags",
       "https://www.gunasthebrand.com/collections/cross-body",
       "https://www.gunasthebrand.com/collections/totes",
       "https://www.gunasthebrand.com/collections/backpacks"]
```

```python
links = []
desc = []
prices = []

for url in urls:
    page = requests.get(url)
    soup = BeautifulSoup(page.content, 'lxml')
    match = soup.findAll('div', {'class':'prod-image'}) # find all tags with images
    links.extend(['http:'+ x.a.img['src'] for x in match])
    desc.extend([x.a['title'] for x in match])
    match2 = soup.findAll('div', {'class': 'price'})
    prices.extend([x.text for x in match2])
```

```python
# create dictionary with all item details to put into dataframe
all_data = []

for i in range(len(links)):
    all_data.append({'description': desc[i], 
                     'link': links[i], 
                     'price': prices[i], 
                    'brand': 'Gunas',
                    'num': i})
```

```python
# list of dictionary in dataframe
df_data = pd.DataFrame(all_data)
df_data
```

```python
df_data.drop_duplicates(subset='description', inplace = True)
df_data
```

```python
df_data.reset_index(inplace=True)
```

```python
df_data
```

```python
df_data.drop(columns = ['index', 'num'], inplace=True)
```

```python
df_data
```

```python
nums = list(range(len(df_data)))
nums
```

```python
df_data = pd.concat([df_data, pd.Series(nums)], axis=1)
```

```python
df_data
```

```python
df_data['img-file'] = df_data[0].apply(lambda x: 'gunas_'+str(int(x)+1)+'.png')
```

```python
df_data
```

```python
df_data.drop(columns = [0], axis = 1, inplace = True)
df_data
```

```python
df_data.price = df_data.price.apply(lambda x: x[-12:].replace("USD", ""))
df_data.price = df_data.price.apply(lambda x: float(x[-12:].replace("\n", "").strip()))
df_data
```

```python
# save the dataframe as a csv (to merge with other data later)
df_data.to_csv('gunas_data.csv')
```

### 2. Loop through image links and save as a .png file

```python
# loop through links to request and save image locally
for i, url in enumerate(df_data.link):
    time.sleep(1)
    try:
        request = urllib.request.Request(url)
        response = urllib.request.urlopen(request)
        binary_str = response.read()
        byte_array = bytearray(binary_str)
        numpy_array = np.asarray(byte_array, dtype='int8')
        image = cv2.imdecode(numpy_array, cv2.IMREAD_UNCHANGED)
        cv2.imwrite("Gunas/"+"gunas_{}".format(i+1)+".png", image)
        print("Saved "+"gunas_{}".format(i+1)+".png")
    except Exception as e:
        print(str(e))

print("Done")
```
